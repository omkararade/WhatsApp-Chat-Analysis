```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from urlextract import URLExtract
from collections import Counter
import emoji
```

```python
f = open('WhatsApp Chat with üòàüíÄCriminal_DiaryüíÄüòà.txt','r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_name' : messages , 'date' : dates})
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'], format = '%d/%m/%Y, %H:%M - ')
```

```python
df.head()
```

```python
df.shape
```

```python
user = []
message = []

for messages in df['user_name']:
    entry = re.split('([\w\W]+?):\s', messages)
    if entry[1:]: # user name
        user.append(entry[1])
        message.append(entry[2])
    else:
        user.append('group_notification')
        message.append(entry[0])

df['user'] = user
df['message'] = message
```

```python
df.head()
```

```python
df.drop(columns=['user_name'],inplace = True)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['month_num'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python

period = []
for hour in df[['day_name','hour']]['hour']:
    if hour ==23:
        period.append(str(hour) + '-' + str('00'))
    elif hour == 0:
        period.append(str('00') + '-' + str(hour+1))
    else:
        period.append(str(hour) + '-' + str(hour+1))
```

```python
df['period'] = period
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name',columns='period',values = 'message',aggfunc = 'count').fillna(0))
plt.yticks(rotation =45)
plt.show()
```
